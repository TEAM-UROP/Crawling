```python
import pandas as pd
import re

mbti_dic = {
    'ì—¥': 'EN', 'ë¿Œ': 'F', 'ì‚': 'P', 'ì—”': 'EN', 'ì¸': 'IN',
    'í‹°': 'T', 'ì œ': 'J', 'ì‡': 'IS', 'ì—£': 'ES', 'íŒ': 'TP',
    'í”„': 'F', 'í‘¸': 'F', '1': 'I', '2': 'E', 'í”¼':'P', 'ì›¬':'EN',
    'ì¤´': 'J', 'ì€¼':'F'
    }

post_csv_file_path = "D:\\ê°•ì˜\\3-2\\í•™ë¶€ì—°êµ¬ì°¸ì—¬(UROP)\\YongRullNewTown\\Crawling\\Post\\post_2023-10-29_14-15-34_(1 to 3).csv"
post_df = pd.read_csv(post_csv_file_path)

def mbti_conversion(row):
    # ì˜ì–´ MBTIë¡œ ë³€í™˜ ê°€ëŠ¥í•œ ê²½ìš°
    mbti_types = re.findall(r'[A-Z]+', row)
    if mbti_types:
        return mbti_types[0]
    
    # í•œê¸€ MBTIë¡œ ë³€í™˜
    converted_string = ''
    for char in row:
        if char in mbti_dic:
            converted_string += mbti_dic[char]
    
    return converted_string

# 'author' ì—´ì„ ë³µì‚¬í•˜ì—¬ 'mbti' ì—´ì„ ìƒì„±
post_df['mbti'] = post_df['author'].apply(mbti_conversion)

# í•„ìš”í•œ MBTI ìœ í˜•ì— í•´ë‹¹í•˜ëŠ” í–‰ë§Œ ì¶”ì¶œ
valid_mbti = ['ISTJ', 'ISFJ', 'INFJ', 'INTJ', 'ISTP', 'ISFP', 'INFP', 'INTP',
              'ESTP', 'ESFP', 'ENFP', 'ENTP', 'ESTJ', 'ESFJ', 'ENFJ', 'ENTJ']
filtered_post_df = post_df[post_df['mbti'].isin(valid_mbti)]

#df ì—´ ìˆœì„œ ë³€ê²½ -> [ë³¸ë¬¸] [ëŒ“ê¸€] [ì‘ì„±ì] [mbti] ìˆœìœ¼ë¡œ
new_order = ['contents', 'comments','author','mbti']
filtered_df = filtered_post_df.reindex(columns=new_order)

# ê²°ê³¼ë¥¼ ì¶œë ¥
print(filtered_post_df)
print(len(filtered_post_df))
print(f'ì •ë‹µë¥  :{(len(filtered_post_df) / len(post_df)) * 100:.2f}%')
```
