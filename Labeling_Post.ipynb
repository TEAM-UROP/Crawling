```python
import pandas as pd
import re

mbti_dic = {
    '엥': 'EN', '뿌': 'F', '삐': 'P', '엔': 'EN', '인': 'IN',
    '티': 'T', '제': 'J', '잇': 'IS', '엣': 'ES', '팁': 'TP',
    '프': 'F', '푸': 'F', '1': 'I', '2': 'E', '피':'P', '웬':'EN',
    '줴': 'J', '쀼':'F'
    }

post_csv_file_path = "D:\\강의\\3-2\\학부연구참여(UROP)\\YongRullNewTown\\Crawling\\Post\\post_2023-10-29_14-15-34_(1 to 3).csv"
post_df = pd.read_csv(post_csv_file_path)

def mbti_conversion(row):
    # 영어 MBTI로 변환 가능한 경우
    mbti_types = re.findall(r'[A-Z]+', row)
    if mbti_types:
        return mbti_types[0]
    
    # 한글 MBTI로 변환
    converted_string = ''
    for char in row:
        if char in mbti_dic:
            converted_string += mbti_dic[char]
    
    return converted_string

# 'author' 열을 복사하여 'mbti' 열을 생성
post_df['mbti'] = post_df['author'].apply(mbti_conversion)

# 필요한 MBTI 유형에 해당하는 행만 추출
valid_mbti = ['ISTJ', 'ISFJ', 'INFJ', 'INTJ', 'ISTP', 'ISFP', 'INFP', 'INTP',
              'ESTP', 'ESFP', 'ENFP', 'ENTP', 'ESTJ', 'ESFJ', 'ENFJ', 'ENTJ']
filtered_post_df = post_df[post_df['mbti'].isin(valid_mbti)]

#df 열 순서 변경 -> [본문] [댓글] [작성자] [mbti] 순으로
new_order = ['contents', 'comments','author','mbti']
filtered_df = filtered_post_df.reindex(columns=new_order)

# 결과를 출력
print(filtered_post_df)
print(len(filtered_post_df))
print(f'정답률 :{(len(filtered_post_df) / len(post_df)) * 100:.2f}%')
```
