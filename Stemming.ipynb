```python
import pandas as pd
import re
from datetime import datetime
from glob import glob
from konlpy.tag import Kkma, Hannanum, Okt
import jpype
import sys
```

```python
sample_comment = pd.read_csv('./tokenized_0.csv')
```

```python
kkma = Kkma()
Hannanum = Hannanum()
Okt = Okt()
```

```python
sample_comment['comments'] = sample_comment['comments'].astype(str)
```

```python
comments_column = sample_comment['comments'].apply(lambda row: ' '.join(kkma.morphs(row)))
```

```python
import pandas as pd
from konlpy.tag import Okt

# Okt 형태소 분석기 객체 생성
okt = Okt()

# 데이터프레임을 받아서 정규화, 어간 추출, 어절 토큰화를 수행하는 함수
def process_dataframe(df, column_name):
    # 정규화, 어간 추출, 어절 토큰화를 수행하는 함수
    def process_text(text):
        normalized_text = okt.normalize(text)
        stemmed_text = ' '.join(okt.morphs(normalized_text, stem=True))
        tokenized_text = ' '.join(okt.morphs(normalized_text, stem=False, norm=False))
        return normalized_text, stemmed_text, tokenized_text

    # 지정된 열에 대해 각각의 처리를 수행하고 새로운 열에 저장
    new_columns = [f'{column_name}_normalized', f'{column_name}_stemmed', f'{column_name}_tokenized']
    df[new_columns] = df[column_name].apply(process_text).apply(pd.Series)

    return df

# 예제 데이터프레임
sample_df = pd.DataFrame(comments_column)

# 함수 호출하여 처리 수행
processed_df = process_dataframe(sample_df, 'comments')

# 결과 출력
processed_df
stem_df = processed_df['comments_stemmed']
```
